```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv(
    '/ecommerce_preparados.csv')
print(df.head())
```

```python
# gráfico que exibe distribuição das notas
plt.figure(figsize=(10, 6))
plt.hist(df['Nota'], bins=20, color='#9AC2C5')
plt.title('Histograma | Distribuição das notas')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.grid(True)
```

```python
# gráfico de dispersão Nota - Quantidade Vendida
plt.figure(figsize=(10, 6))
plt.scatter(df['Nota'], df['Nota'])
plt.title('Dispersão | Nota x Nota')
plt.xlabel('Nota')
plt.ylabel('Nota')
```

```python
# mapa de Calor
plt.figure(figsize=(10, 6))
corr = df[['Nota', 'N_Avaliações']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação Nota x  Numero de avaliações')
```

```python
# Gráfico de Barras Qtd Vendidos
plt.figure(figsize=(10, 6))
df['Nota'].value_counts().plot(kind='bar', color= '#CAA8F5')
plt.xlabel('Nota')
plt.ylabel('Frequência')
```

```python
# Gráfico de Pizza
x = df['Temporada_Cod'].value_counts().index
y = df['Temporada_Cod'].value_counts().values
plt.figure(figsize=(8, 8))
plt.pie(
    y,
    labels=x,
    autopct='%.1f%%',
    startangle=0,
)
plt.title('Distribuição por Temporada')
```

```python
#Gráfico de Densidade
plt.figure(figsize=(10,6))
sns.kdeplot(df['Nota'], fill=True, color='#9984D4')
plt.title('Densidade de Notas')
plt.xlabel('Notas')
```

### Operadores Relacionais

```python
#Gráfico de regressão
sns.regplot(x='Nota',
            y='Nota_MinMax',
            data=df, color='#9984D4',
            scatter_kws={'alpha':  0.5, 'color': '#592E83'})
plt.title('Regressão de Nota')
plt.xlabel('Nota')
plt.ylabel('Nota_MinMax')
```
